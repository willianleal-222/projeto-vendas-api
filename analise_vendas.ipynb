```python
import pandas as pd
```

```python
df = pd.read_csv("vendas.csv")
df
```

```python
df.info()
```

```python
df['faturamento'] = df['valor'] * df['quantidade']
df
```

```python
df['faturamento'].sum()
```

```python
df.groupby('produto')['quantidade'].sum()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df['mes'] = df['data'].dt.month
df.groupby('mes')['faturamento'].sum()
```

```python
faturamento_total = df['faturamento'].sum()
top_produtos = df.groupby('produto')['quantidade'].sum()
vendas_mes = df.groupby('mes')['faturamento'].sum()

print("Faturamento total:", faturamento_total)
print("\nTop produtos:\n", top_produtos)
print("\nVendas por mês:\n", vendas_mes)
```
